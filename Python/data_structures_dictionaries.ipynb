# What is a Dictionary?

A dictionary stores data as key → value pairs.

##Key Properties

- Unordered (logically)

- Mutable

- Keys must be unique

- Keys must be immutable (str, int, tuple)

- Values can be any type

###`Dictionary = real-world record / object`

##Why Dictionaries Matter in AI / ML

Used for:

- Dataset rows

- JSON data

- Model configs

- Hyperparameters

- Feature-value mapping

- Label encoding

## Creating Dictionaries

```python
person = {"name": "Ali", "age": 21,"degree":"CS","batch":23}

# using dict
# person = dict(name="Ali", age=21)
# empty
data = {}
```

## Accessing Values

```python
print(person["name"])

# print(person["city"])  # KeyError if key doesn't exist

# Safe access - get()
print(person.get("city"))        # None
print(person.get("city", "N/A")) # Default value

# Adding & Updating Values
person["city"] = "Lahore"   # add
person["age"] = 22         # update

```

## Removing Items

```python
print(person)
# pop
# Returns the value associated with the removed key
# Can return a default value if specified, preventing an error
age = person.pop("age",None)
print(person)


# popitem() (removes last inserted pair)
person.popitem()
print(person)

# del
# Does not return a value
# Always raises a KeyError if the key does not exist
# del person["city"]
print(person)

# clear the dictionery
person.clear()

# length of dict
print(len(person))
print(person)

```

## Looping Through Dictionaries

```python
for key in person:
    print(key)

for value in person.values():
    print(value)

for key, value in person.items():
    print(key, value)
```

###What is a Shallow Copy?

A shallow copy creates a new outer object, but inner objects are shared.

Outer container is copied
Inner data is NOT copied

```python
original = {
    "name": "Ali",
    "marks": [80, 90]
}

copy_dict = original.copy()
# original  ──►  marks list ◄──  copy_dict
# original and copy_dict are pointing to same marks list in memory

# modifying inner structure data
copy_dict["marks"].append(100)

print(original)
print(copy_dict)

# modifying outer structure data
copy_dict["name"] = "Sara"

print(original)
print(copy_dict)

```

```python
# update method
"""
New Keys: If a key from other doesn't exist in the original, it is added.
Existing Keys: If a key already exists, its value is overwritten by the new value.
Return Value: It returns None and modifies the dictionary in-place

"""
profile = {"name": "Bob", "status": "active"}
new_data = {"status": "away", "city": "New York"}

profile.update(new_data)
print(profile)
# Output: {'name': 'Bob', 'status': 'away', 'city': 'New York'}
```

```python
# setdefault()
"""
If the key exists: It returns the existing value and does not change the dictionary.
If the key is missing: It inserts the key with a specified default value and returns that value.
Default parameter: If you don't provide a second argument, it defaults to None
"""
data = [("fruit", "apple"), ("veg", "carrot"), ("fruit", "banana")]
groups = {}

for category, item in data:
    # 1. Look for 'category'. If missing, set it to []
    # 2. Return that list (either existing or new)
    # 3. Append the item directly to it
    groups.setdefault(category, []).append(item)

print(groups)
# Output: {'fruit': ['apple', 'banana'], 'veg': ['carrot']}

scores = {}
scores.setdefault("Ali", 0)
print(scores)
scores["Ali"] += 10
print(scores)
```

```python
# Nested Dictionaries
students = {
    "s1": {"name": "Ali", "marks": 85},
    "s2": {"name": "Sara", "marks": 90}
}
print(students)
print(students["s1"]["marks"])
```

## Dictionary Methods
| Method | Purpose |
| :--- | :--- |
| `get()` | Safely access a value without raising a KeyError |
| `keys()` | Returns a view object containing all dictionary keys |
| `values()` | Returns a view object containing all dictionary values |
| `items()` | Returns a view object containing all key-value tuples |
| `update()` | Merges another dictionary or iterable into the current one |
| `pop()` | Removes a specific key and returns its value |
| `popitem()` | Removes and returns the last inserted key-value pair |
| `clear()` | Removes all elements from the dictionary |
| `copy()` | Returns a shallow copy of the dictionary |
| `setdefault()` | Returns a key's value; inserts the key if it doesn't exist |


```python
squares = {x: x*x for x in range(5)}
print(squares)

# Using Tuples as Keys
coordinates = {(0,0): "origin", (1,2): "point"} # using list as keys is error
print(coordinates)
```
