```python
# Step 1: Import the pipeline
from transformers import pipeline

# Step 2: Load the Whisper model
# We use "openai/whisper-tiny" again
transcriber = pipeline("automatic-speech-recognition", model="openai/whisper-tiny")

# Step 3: A Reliable Audio URL (Martin Luther King Jr. clip)
# .flac is a high-quality audio format often used in ML
audio_url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"

# Step 4: Transcribe
result = transcriber(audio_url)

print("-" * 50)
print("Transcription:", result['text'])
print("-" * 50)
```

```python
# Step 1: Install the necessary library (if not already installed)
# !pip install sentence-transformers

from sentence_transformers import SentenceTransformer, util

# Step 2: Load the Model
# "all-MiniLM-L6-v2" is a tiny, super-fast model designed for this exact job
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 3: The Client's Database (List of products)
products = ["Nike Air Max", "Apple iPhone 14", "Samsung Galaxy S23", "Organic Bananas", "Adidas Running Shoes"]

# Step 4: The User's Search Query
# Notice the user uses "cellphone" (which does NOT appear in the list above)
query = "cellphone"

# Step 5: Encode (Turn text into Numbers/Vectors)
product_embeddings = model.encode(products)
query_embedding = model.encode(query)

# Step 6: Find the Best Match (Cosine Similarity)
# We compare the query numbers against the product numbers
scores = util.cos_sim(query_embedding, product_embeddings)[0]

# Step 7: Print Results
print(f"Query: {query}\n")
# Combine products with their scores and sort them
results = sorted(zip(products, scores), key=lambda x: x[1], reverse=True)

for product, score in results:
    print(f"Product: {product}  |  Match Score: {round(score.item(), 2)}")
```

```python
# Step 1: Import the pipeline
from transformers import pipeline

# Step 2: Load the Object Detection pipeline
# This downloads 'facebook/detr-resnet-50'
detector = pipeline("object-detection")

# Step 3: The Client's Image
# A busy image with people, bicycles, and cars
url = "http://images.cocodataset.org/val2017/000000039769.jpg" # (Using our cat image again to see where they are)

# Step 4: Detect!
results = detector(url)

# Step 5: Print what it found and WHERE it found it
for item in results:
    label = item['label']
    score = round(item['score'], 2)
    box = item['box'] # The coordinates [xmin, ymin, xmax, ymax]

    print(f"Found: {label} | Confidence: {score}")
    print(f"   Location: {box}")
    print("-" * 30)
```

```python
# Step 1: Install specific audio libraries and espeak-ng
# !sudo apt-get install -y espeak-ng
# !pip install datasets soundfile phonemizer

from transformers import pipeline
from IPython.display import Audio # To play sound in the notebook

# Step 2: Load the TTS pipeline
# "kakao-enterprise/vits-ljs" is a great English voice model
synthesizer = pipeline("text-to-speech", model="kakao-enterprise/vits-ljs")

# Step 3: The Text to Speak
text = "hehehehehehhehehehheh"

# Step 4: Generate Audio
result = synthesizer(text)

# Step 5: Play the Audio
# The audio data is in result['audio'][0]
audio_data = result['audio'][0]
sampling_rate = result['sampling_rate']

Audio(data=audio_data, rate=sampling_rate)
```

```python
# --- STEP 1: INSTALLS & SETUP ---
# We keep the installs just in case
!sudo apt-get install -y espeak-ng > /dev/null 2>&1
!sudo apt-get install -y ffmpeg
!pip install git+https://github.com/huggingface/transformers.git accelerate datasets soundfile librosa gradio phonemizer

import gradio as gr
from transformers import pipeline

# --- STEP 2: LOAD THE "ORGANS" ---

print("1. Loading the Ear (Whisper)...")
ear = pipeline("automatic-speech-recognition", model="openai/whisper-tiny")

print("2. Loading the Brain (DistilGPT2)...")
brain = pipeline("text-generation", model="distilgpt2")

print("3. Loading the Mouth (MMS)...")
# SWITCHED MODEL: "facebook/mms-tts-eng" is simpler and more stable
mouth = pipeline("text-to-speech", model="facebook/mms-tts-eng")

# --- STEP 3: THE "THINKING" FUNCTION ---

# --- DEBUG VERSION OF MINI-SIRI ---

def mini_siri(audio_path):
    # Print the path to prove the file actually arrived
    print(f"DEBUG: Audio file received at {audio_path}")

    if audio_path is None:
        return "No audio found", "Please upload a file", None

    # 1. HEAR (Removing the try-except block so we can see real errors)
    # This might crash if ffmpeg is missing
    hearing_results = ear(audio_path)
    user_text = hearing_results["text"]
    print(f"DEBUG: User said: {user_text}")

    # 2. THINK
    input_text = f"Question: {user_text} Answer:"
    thought_results = brain(input_text, max_length=50, num_return_sequences=1)
    ai_text = thought_results[0]['generated_text'].replace(input_text, "").strip()

    if len(ai_text) == 0:
        ai_text = "I am sorry, I did not understand."

    # 3. SPEAK
    speech = mouth(ai_text)

    return user_text, ai_text, (speech["sampling_rate"], speech["audio"][0])

# Re-launch the interface with the new function
interface = gr.Interface(
    fn=mini_siri,
    inputs=gr.Audio(sources=["upload", "microphone"], type="filepath"),
    outputs=[
        gr.Textbox(label="You Said"),
        gr.Textbox(label="AI Said"),
        gr.Audio(label="AI Voice")
    ],
    title="Mini-Siri (Debug Mode) 🔧",
    description="Upload an audio file. If it crashes, copy the RED error message and show me!"
)

interface.launch(debug=True)
```
