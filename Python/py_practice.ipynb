#Module 2 Grand Assignment: The "Step"
####NeuronContext:  
In the 1950s, the first AI (the Perceptron) used a Step Activation Function. It worked like this:  
Calculate $(x \cdot w) + b$.  
If the result is greater than 0, the neuron fires (output 1).  
If the result is 0 or less, the neuron stays silent (output 0).  
Your Mission:Write a program that simulates a Perceptron Training Loop.  Setup:Create a variable input_x = 0.5  
Create a variable weight = 0.5  
Create a variable bias = -1.0 (notice it's negative).  
The Loop:Create a for loop that runs for 10 epochs (range 10).  
Inside the Loop:  
Calculate output = (input_x * weight) + bias.  
The Activation Check: Use an if/else statement.  

If output > 0: Print "Epoch [X]: Neuron FIRED (Output: [output])"If output <= 0: Print "Epoch [X]: Neuron DORMANT (Output: [output])"The Learning Step:After printing, increase the bias by 0.2. (This simulates the AI "learning" and adjusting its threshold).The Goal:You should see the neuron stay "DORMANT" for the first few epochs, and then switch to "FIRED" as the bias increases.

```python
# --- Perceptron Training Loop Simulator ---

# 1. Setup (Initialization)
input_x = 0.5
weight = 0.5
bias = -1.0  # We start with a strong negative bias

print("--- STARTING TRAINING SESSION ---")

# 2. The Training Loop (10 Epochs)
for epoch in range(10):

    # A. Forward Pass
    output = (input_x * weight) + bias

    # B. Activation Function (The "Step" Logic)
    # This decides if the neuron fires (1) or stays silent (0)
    if output > 0:
        state = "FIRED"
    else:
        state = "DORMANT"

    # Print the status for this epoch
    # We use round() to keep the output readable
    print(f"Epoch {epoch+1}: Neuron {state} (Output: {round(output, 2)})")

    # C. The Learning Step (Parameter Update)
    # We slowly increase the bias to make the neuron more sensitive
    bias = bias + 0.2

print("--- TRAINING COMPLETE ---")
```

## Module 4 Grand Assignment: The "Hyperparameter Tuner"
This assignment requires you to combine defaults, **kwargs, type hinting, and return values.

Task: You are building a function that configures a Neural Network.

Requirements:

Define a function named configure_model.

Parameters:

model_name (str) - Mandatory.

learning_rate (float) - Optional, default 0.001.

**extra_params - Captures any other settings.

Inside the function:

Print "Configuring [model_name]..."

Check if learning_rate is greater than 0.1. If yes, print "Warning: High learning rate!"

Print "Extra Parameters:" and then print the extra_params dictionary.

Return a dictionary containing all the settings combined (Name, LR, and the extras).

Execution: Call the function with these arguments:

Name: "VisionNet"

Learning Rate: 0.5

Extra arguments: optimizer="Adam", epochs=50, use_gpu=True.

Output: Store the returned dictionary in a variable called final_config and print it at the end.

```python
# --- AI Model Configurator ---

# 1. Function Definition with Type Hinting
# We specify that this function returns a 'dict' (-> dict)
def configure_model(model_name: str, learning_rate: float = 0.001, **extra_params) -> dict:

    # A. Status Update
    print(f"Configuring {model_name}...")

    # B. Logic Check (Safety Mechanism)
    if learning_rate > 0.1:
        print("Warning: High learning rate! Training might be unstable.")

    # C. Inspecting the Flexible Arguments (**kwargs)
    # Python automatically packs 'optimizer', 'epochs', etc. into this dictionary.
    print(f"Extra Parameters: {extra_params}")

    # D. Constructing the Return Object
    # We start with the known arguments
    full_config = {
        "Model": model_name,
        "LR": learning_rate
    }

    # We merge the 'extra_params' dictionary into our main dictionary
    # .update() adds all key-value pairs from one dict to another
    full_config.update(extra_params)

    # E. Sending Data Back
    return full_config

# 2. Execution Phase
print("--- SYSTEM START ---")

# We call the function with a mix of positional, keyword, and 'extra' arguments
final_config = configure_model(
    "VisionNet",
    learning_rate=0.5,     # Overriding default (triggers warning)
    optimizer="Adam",      # Captured by **kwargs
    epochs=50,             # Captured by **kwargs
    use_gpu=True           # Captured by **kwargs
)

# 3. Verification
print("\n--- FINAL CONFIG RETURNED ---")
print(final_config)
```

2. Generators & The yield Keyword (Critical for Big Data)  
This is how we handle massive datasets. A standard function returns data and kills the function. A generator yields data one piece at a time and pauses the function, remembering where it left off.

Memory Impact:

List: Loads 1,000,000 images into RAM at once. (Crash)

Generator: Loads 1 image, gives it to the AI, discards it, loads the next. (Safe)

```python
def data_loader(limit):
    for i in range(limit):
        # ... Imagine complex code here to load an image ...
        yield f"Image_{i}.jpg"
        # The function PAUSES here until asked for the next item

# Usage
loader = data_loader(5)

# We get data on demand using next()
print(next(loader)) # Output: Image_0.jpg
print(next(loader)) # Output: Image_1.jpg
```

```python
# This generator function opens the file and yields lines one by one
def read_massive_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line  # Sends one line to the loop and PAUSES

# Using the generator
for line in read_massive_file("massive_log_file.txt"):
    if "ERROR" in line:
        print(f"Found an error: {line.strip()}")
    # After each loop, Python goes back to 'yield' and gets the NEXT line
```

AI training takes hours. You don't want a minor error (like a corrupted file) to stop the entire process at 99%.

try: Run this code.

finally: Always run this at the end (even if it crashes).

The "Finally" trick: In AI, we use finally to save our model's progress. If the code crashes or you stop it with Ctrl+C, the finally block ensures you don't lose your work.

```python
try:
    x = 10 / 0  # This causes ZeroDivisionError
except ZeroDivisionError as e:
    print(f"Caught an error: {e}")
finally:
    print("Saving Checkpoint... (This runs no matter what)")
```

3. Encapsulation (Hiding Data)  
In AI, you don't want users messing with internal weights manually. You "hide" them using Private Attributes (prefix with __ double underscore).

```python
class NeuralNetwork:
    def __init__(self):
        self.layers = 2       # Public (Everyone can see)
        self.__weights = 0.5  # Private (Hidden)

    def get_weights(self):
        return self.__weights  # Safe access method

model = NeuralNetwork()
print(model.layers)     # Works
# print(model.__weights) # CRASH! Python says "AttributeError: No such attribute"
print(model.get_weights()) # Works
```

4. Inheritance (The Power of Reusability)  
This is essential for Deep Learning. You create a "Base Class" (generic) and then create "Child Classes" (specific) that inherit all the code from the parent.

```python
# Parent Class
class Model:
    def save(self):
        print("Saving model to disk...")

# Child Class inherits from Model
class VisionModel(Model):
    def predict(self):
        print("Detecting objects...")

# Usage
my_ai = VisionModel()
my_ai.predict() # Defined in Child
my_ai.save()    # Defined in Parent (It works!)
```

##Module 6 Grand Assignment: The "Agent" Factory  
Task: You are building the foundation for an Agentic AI system. You need a blueprint for generic Agents and a specific blueprint for a "Chat Agent."

Requirements:

Parent Class (Agent):

Define a class named Agent.

__init__: Accepts name and status (default to "Idle"). Store them as attributes.

Method: work(). It should print "[Name] is working...".

Child Class (ChatAgent):

Create a class ChatAgent that inherits from Agent.

__init__:

Accepts name.

Call the parent's __init__ using super().__init__(name). (Look up super() if needed, or I will explain in the solution. It calls the parent's setup code).

Initialize a new attribute context_length = 4096.

Method Override: Redefine work(). It should print "[Name] is generating text with context [4096]...".

Execution:

Create a generic Agent named "Worker_1".

Create a ChatAgent named "GPT_Bot".

Call .work() on both.

Verify that "Worker_1" does the generic work, and "GPT_Bot" does the specific text work.

Go. This structure (Base Class -> Specific Class) is exactly how PyTorch Layers and Models are built.

```python
# --- Agentic AI Factory ---

# 1. The Parent Class (The Blueprint)
class Agent:
    def __init__(self, name, status="Idle"):
        self.name = name
        self.status = status
        print(f"[System] Created Base Agent: {self.name}")

    def work(self):
        # Generic behavior
        print(f"{self.name} is performing generic tasks...")

# 2. The Child Class (The Specialist)
# "class Child(Parent)" means ChatAgent INHERITS from Agent
class ChatAgent(Agent):
    def __init__(self, name):
        # SUPER() - The Critical Link
        # We call the Parent's __init__ to handle the name/status setup
        super().__init__(name, status="Active")

        # Then we add specific attributes for this child
        self.context_length = 4096

    # METHOD OVERRIDING
    # We replace the parent's generic 'work' with a specific version
    def work(self):
        print(f"{self.name} is generating text with context length [{self.context_length}]...")

# 3. Execution Phase
print("--- INITIALIZING AGENTS ---")
worker_1 = Agent("Worker_1")
gpt_bot = ChatAgent("GPT_Bot")

print("\n--- AGENT WORKFLOW ---")
worker_1.work()  # Runs the Parent's code
gpt_bot.work()   # Runs the Child's OVERRIDDEN code
```
