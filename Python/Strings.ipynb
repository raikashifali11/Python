##What is a String?

A string is a sequence of characters enclosed in quotes

In Python:

- Strings are immutable

- They are iterables

- They are Unicode by default

```python
# Creating Strings
# Single Quotes
s = 'hello'

# Double Quotes
s = "hello"

# Triple Quotes (Multiline)
s = """This
is
multi-line"""
```

```python
# String Immutability

# Strings cannot be changed in-place

s = "ai"
# s[0] = "A"   # ERROR


# You must create a new string:

s = "A" + s[1:]
print(s)
```

```python
# Indexing & Slicing
# Indexing
s = "python"
print(s[0])   # 'p'
print(s[-1])   # 'n'

# Slicing
# Slicing follows the syntax [start:stop:step] default is [0,len,1]
print(s[0:3])    # 'pyt'
print(s[:4])    # 'pyth'
print(s[::2])    # 'pto'
print(s[::-1])    # reverse
```

```python
# Length & Iteration
len(s)


# Iterating:

for char in s:
    print(char)
```

```python
# Membership Operators
"ai" in "machine learning"     # True
"z" not in s                   # True
```

```python
# String Concatenation & Repetition
"a" + "b"
"ha" * 3
```

```python
texts = [
    "   AI is THE future!!! 🤖🤖   ",
    "Machine   Learning is AMAZING!!!",
    "Python_is_great_for_AI ",
    "   NLP is fun, powerful, and useful   "
]
print(texts)
```

```python
# Final processed data for each text
processed_results = []

# Vocabulary set (unique words)
vocabulary = set()

# Token frequency dictionary
token_frequency = {}
```

```python
for text in texts:

    clean_text = text.lower().strip()

    clean_text = clean_text.replace("!","")
    clean_text = clean_text.replace("_"," ")
    clean_text = clean_text.replace(",","")

    clean_text = " ".join(clean_text.split())

    # Discards any characters that are not valid UTF-8
    clean_text = clean_text.encode("utf-8", errors="ignore").decode("utf-8")

    tokens = clean_text.split()

    filtered_tokens = []
    for token in tokens:
        if len(token) >= 3 and not token.isdigit():
            filtered_tokens.append(token)

    for token in tokens:
        vocabulary.add(token)
        token_frequency[token] = token_frequency.get(token,0) + 1

    processed_results.append({
    "original_text": text,
    "clean_text": clean_text,
    "tokens": filtered_tokens,
    "token_count": len(filtered_tokens)
    })
```

```python
# Display structured results
for result in processed_results:
    print("Original Text:", result["original_text"])
    print("Clean Text   :", result["clean_text"])
    print("Tokens       :", result["tokens"])
    print("Token Count  :", result["token_count"])
    print("-" * 50)
```

```python
# Sort frequency dictionary by count (descending)
top_words = sorted(
    token_frequency.items(),
    key=lambda item: item[1],
    reverse=True
)

print("Top 5 Most Frequent Words:")
for word, count in top_words[:5]:
    print(word, "→", count)
```

##Breakdown of the Code for sort frequency

###token_frequency.items():  

Converts the dictionary into a list of tuples.  
Example: {"apple": 5, "banana": 2} becomes [("apple", 5), ("banana", 2)].

###key=lambda item: item[1]:  

This tells the sorted() function how to compare the tuples.  
item represents one tuple (e.g., ("apple", 5)).  
item[1] points to the count (the value). This ensures we sort by the frequency, not by the word's name.  

###reverse=True:  

By default, Python sorts in ascending order (smallest to largest).  
reverse=True flips this so the highest counts appear first
