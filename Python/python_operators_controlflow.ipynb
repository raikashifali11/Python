## Operators in Python

Operators are special symbols used to perform operations on values.


## Arithmetic Operators

| Operator | Meaning |
|--------|--------|
| + | Addition |
| - | Subtraction |
| * | Multiplication |
| / | Division |
| % | Modulus (remainder) |
| // | Floor division |
| ** | Power |


````python
# Arithmetic operators example

a = 11
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Floor Division:", a // b)
print("Power:", a ** b)
````

## Comparison Operators

Used to compare two values.
Result is always **True or False**.

| Operator | Meaning |
|--------|--------|
| == | Equal |
| != | Not equal |
| > | Greater than |
| < | Less than |
| >= | Greater than or equal |
| <= | Less than or equal |


````python
x = 5
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
````

## Logical Operators

Used to combine conditions.

| Operator | Meaning |
|--------|--------|
| and | Both conditions must be True |
| or | At least one condition True |
| not | Reverse the result |


````python
age = 20
has_id = True

print(age >= 18 and has_id)
print(age < 18 or has_id)
print(not has_id)
````

## Assignment Operators

Used to assign and update values.


````python
x = 10

x += 5   # x = x + 5
x -= 3   # x = x - 3
x *= 2   # x = x * 2

print(x)
````

# Control Flow – Decision Making

Control flow allows Python to make decisions using conditions.


## if Statement

Syntax:
```python
if condition:
    statement


````python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote")
````

## if-else Statement


````python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")

if number % 5 == 0:
    print("Divisble by 5")
else:
    print("Not divisble by 5")
````

## if-elif-else Statement


````python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
````

# Loops in Python

Loops are used to repeat a block of code.


## for Loop


````python
# for loop example

#for i in range(1, 6):
#    print("Number:", i)
for i in range(1,21):
    print("7 x ",i," = ", i*7)
````

## while Loop


````python
count = 1

while count <= 100:
    print("Count:", count)
    count += 2
````

## break, continue, pass

- break → stops loop
- continue → skips current iteration
- pass → does nothing


````python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
````

````python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
````

````python
"""The pass keyword in Python is a null operation used
   as a placeholder where a statement is syntactically required,
   but no action or code needs to be executed """

for i in range(1, 4):
    if i == 2:
        pass
    print(i)
````

## Other IMP Advanced concepts

````python

# ---------- BITWISE OPERATORS ----------
a = 10      # 1010 (binary)
b = 4       # 0100 (binary)

print(a & b)    # AND: 0000 -> 0
print(a | b)    # OR : 1110 -> 14
print(a ^ b)    # XOR: 1110 -> 14
print(~a)       # NOT: invert bits (two's complement) --> -(a+1) = -(10+1) = 11
print(a << 1)   # Left shift (multiply by 2)
print(a >> 1)   # Right shift (divide by 2)

````

````python
# ---------- MEMBERSHIP OPERATORS ----------
nums = [1, 2, 3, 4, 5]

print(3 in nums)        # True if element exists
print(10 not in nums)  # True if element does not exist
````

````python
# ---------- IDENTITY OPERATORS ----------
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)       # True: same memory reference
print(x is z)       # False: different objects
print(x == z)       # True: values are equal


# ---------- OPERATOR PRECEDENCE ----------
result = 10 + 5 * 2       # * has higher precedence
result2 = (10 + 5) * 2    # parentheses override precedence

print(result)
print(result2)
````

````python
# ---------- NESTED IF ----------
num = 15

if num > 0:
    if num % 5 == 0:
        print("Positive and divisible by 5")


# ---------- TERNARY (CONDITIONAL EXPRESSION) ----------
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)


# ---------- CHAINED COMPARISON ----------
x = 7
print(0 < x < 10)   # Equivalent to (x > 0 and x < 10)
````

````python
# ---------- RANGE() ADVANCED ----------
print(list(range(5)))           # 0 to 4
print(list(range(2, 6)))        # 2 to 5
print(list(range(1, 10, 2)))    # Step of 2


# ---------- NESTED LOOPS ----------
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)


# ---------- FOR-ELSE ----------
for n in range(2, 6):
    if n == 7:
        break
else:
    print("Loop completed without break")


# ---------- WHILE-ELSE ----------
count = 0
while count < 3:
    print(count)
    count += 1
else:
    print("While loop ended normally")
````

````python
# ---------- BREAK IN NESTED LOOP ----------
for i in range(1, 4):
    for j in range(1, 4):
        if j == 2:
            break
        print(i, j)


# ---------- CONTINUE WITH CONDITION ----------
for i in range(1, 6):
    if i % 2 == 0:
        continue    # Skip even numbers
    print(i)


# ---------- PASS AS PLACEHOLDER ----------
for i in range(3):
    if i == 1:
        pass        # Used when syntax is required but no action
    print(i)

````

````python
# ---------- BUILT-IN NUMERIC FUNCTIONS ----------
print(abs(-10))          # Absolute value
print(pow(2, 3))         # Power
print(round(3.14159, 2)) # Rounding
print(divmod(10, 3))     # (quotient, remainder)


# ---------- ENUMERATE ----------
names = ["Ali", "Sara", "Ahmed"]
for index, name in enumerate(names):
    print(index, name)

alphabets = ["a","b"]
for index, alphabet in enumerate(alphabets):
    print(index,alphabet)

# ---------- ZIP ---------- using multiple lists as single iterator
stud_names = ["Ali ahmad", "Sana", "Kashif"]
scores = [85, 90, 78]
for name, score in zip(stud_names, scores):
    print(name, score)


# ---------- ANY & ALL ----------
values = [True, False, True]

print(any(values))   # True if at least one True
print(all(values))   # True only if all True


# ---------- isinstance() ----------
x = 10
print(isinstance(x, int))     # True
print(isinstance(x, float))  # False


# ---------- BOOLEAN TRUTHINESS ----------
print(bool(0))       # False
print(bool(1))       # True
print(bool(""))      # False
print(bool("AI"))    # True


# ---------- SHORT-CIRCUIT LOGIC ----------
x = 0
y = 10

# Second condition not evaluated because first is False
print(x != 0 and y / x > 2)

# Second condition not evaluated because first is True
print(x == 0 or y / x > 2)


# ---------- SENTINEL CONTROLLED LOOP ----------
while True:
    value = input("Enter number (q to quit): ")
    if value == 'q':
        break
    print(int(value) * 2)


# ---------- FLAG VARIABLE ----------
found = False
data = [1, 3, 5, 7]

for n in data:
    if n == 5:
        found = True
        break

print(found)
````

````python
# part A
names = []
marks_list = []
attendance_list = []
status_flags = []
while True:
    name = input("Enter your name:(q to quit) ")
    if name == 'q':
        break
    marks = input("Enter your marks (0-100): ")
    attendance = input("Enter your attendance percentage: ")


    if not name:
        print("Name skipped")
        continue

    if not marks.isdigit() or not attendance.isdigit():
        print("Invalid numeric values")
        continue
    marks = int(marks)
    attendance = int(attendance)
    if not (0 <= marks <=100 and 0 <= attendance <=100):
        print("values out of range ")
        continue


    names.append(name)
    marks_list.append(marks)
    attendance_list.append(attendance)
else:
    print("loop ended successfully")
print(names)
print(marks_list)
print(attendance_list)
````

````python
# part B - grade and status logic

for i in range(len(names)):

    marks = marks_list[i]
    attendance = attendance_list[i]

    if marks >= 90:
        grade = 'A'
    elif marks >= 80:
        grade = 'B'
    elif marks >= 70:
        grade = 'C'
    elif marks >= 60:
        grade = 'D'
    else:
        grade = 'F'

    # calculating result using ternary operators
    result = "Pass" if marks >= 50 else "Fail"

    # scholarship eligibility
    scholarship = marks >= 85 and attendance >= 90

    # exam qualification
    disqualified = marks < 50 or attendance < 75

# part C
    squared_marks = pow(marks,2)
    quotient,remainder = divmod(marks,3)

    # Bitwise encoding of status
    # bit 0 → pass/fail
    # bit 1 → scholarship
    # bit 2 → disqualified
    # (status = 1 --> pass) and (status = 3 --> pass+scholarship) and (status = 4 --> disqualified)
    status = 0
    status |= (marks >= 50) << 0
    status |= scholarship << 1
    status |= disqualified << 2

    status_flags.append(status)
    # Assignment operators
    # these lines do not change anything used for syntax practice
    marks_list[i] += 0    # explicit usage
    attendance_list[i] *= 1

````

````python
# part D

for index, (n, m, a, s) in enumerate(zip(names,marks_list,attendance_list,status_flags)):

    # memborship operators
    if n not in names:
        continue
    if s is None:
        pass
    elif s == 0:
        print("No status flags")

    print(f"\nStudent {index+1}")
    print("Name: ",n)
    print("Marks: ",m)
    print("Attendance: ",a)
    print("Status: ",s)
# for-else demonstration
else:
    print("\nAll student records processed")
````
