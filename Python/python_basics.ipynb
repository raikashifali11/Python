# Python Phase 1 – Basics

## What is Python?
Python is a high-level, interpreted programming language.
It is:
- Easy to read
- Easy to write
- Widely used in AI, Data Science, Web, Automation

Python reads code **line by line**, so indentation (spaces) is very important.


## Why Python is Beginner Friendly
- Simple English-like syntax
- No need to declare variable types
- Huge community and libraries
- Works on all platforms (Windows, Linux, Mac)

Python files normally end with `.py`,  
but in Jupyter Notebook we use `.ipynb`.



## Python Syntax & Indentation

Python uses **indentation instead of brackets `{}`**.
Indentation tells Python which code belongs together.

Example:
```python

## Comments in Python

Comments are lines that Python ignores.
They are used to explain code.

Types of comments:
1. Single-line comment → `#`
2. Multi-line comment → triple quotes (""" or ''')


````python
# This is a single-line comment

"""
This is a multi-line comment
Python will ignore this text
"""

print("Comments are not executed")
````

## Variables in Python

A variable is used to store data.

Python does NOT require:
- Type declaration
- Semicolon

Syntax:
variable_name = value


````python
# Creating variables
age = 20
name = "Ali"
height = 5.9
is_student = True

# Printing variables
print(age)
print(name)
print(height)
print(is_student)
````

## Data Types in Python

Common data types:
- int → Integer numbers
- float → Decimal numbers
- str → Text
- bool → True / False


````python
x = 10        # int
y = 3.14      # float
z = "Python"  # string
a = True      # boolean

# Checking data type using type() function
print(type(x))
print(type(y))
print(type(z))
print(type(a))
print(type(y))
````

## Input from User

Python uses `input()` function to take input from user.

Important:
- `input()` always returns **string**


````python
# Taking input from user
user_name = input("Enter your name: ")
user_age = input("Enter your age: ")

# Printing user input
print("Hello", user_name, "your age is",user_age)
````

## Type Casting

Type casting means converting one data type into another.

Common functions:
- int()
- float()
- str()
- bool()


````python
# Taking number input
num1 = input("Enter a number: ")

# Converting string to integer
num1 = int(num1)
num2 = int(num1)

# Performing calculation
result = num1 + num2

print("Result:", result)
````

## Printing Multiple Values

Python allows printing multiple values using commas.


````python
name = "Python"
version = 3.12

print("Language:", name, "Version:", version)
````

# Phase 1 Mega Assignment – Student Profile System

This program collects student information, performs basic calculations,
and displays results using Python basics.


## Requirements

The program must:
1. Take student details from user
2. Store data in variables
3. Display data with proper messages
4. Perform numeric calculations
5. Use type casting
6. Show data types using type()
7. Contain comments explaining code


````python
# Taking details from the user in input
user_name = input("Enter your name: ")
user_age = int(input("Enter your age: "))
user_height = float(input("Enter your height: "))
user_enrollment_status = bool(input("Enter your enrollment status (true/false): "))

# displaying the details of the user
print("\n")
print("Your name is:", user_name, type(user_name))
print("Your age is:", user_age, type(user_age))
print("Your height is:", user_height, type(user_height))
print("Your enrollment status is:", user_enrollment_status, type(user_enrollment_status))
````

````python
# taking two numbers as input
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# performing arithmetic operations on two numbers
sum = num1 + num2
difference = num1 - num2
product = num1 * num2

# displaying the results
print("The sum to two numbers is ",sum)
print("The difference to two numbers is ",difference)
print("The product to two numbers is ",product)
````
