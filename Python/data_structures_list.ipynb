# PART 1: LISTS

 Concept: What is a List?

A list is:

Ordered

Mutable (can be changed)

Allow duplicates

Can store mixed data types

have zero based indexing


```python
numbers = [10,10,10, 20, 30,10]
names = ["Ali", "Sara", "Ahmed"]
mixed = [1, "AI", True, 3.14]

print(numbers[0])   # First element
print(numbers[-1])  # Last element

print(numbers[0:2])   # [10, 20]
print(numbers[:2])    # [10, 20]
print(numbers[1:])    # [20, 30]
print(numbers[::-1])  # Reverse list

# modifying
numbers[1] = 99
print(numbers)

# adding
numbers.append(40)
print(numbers)

# adding multiple
numbers.extend([50,60,70])
print(numbers)

# add at index
numbers.insert(1, 15)
print(numbers)

# remove by value - only first found
numbers.remove(15)
print(numbers)

# pop() – by index (default last)
numbers.pop()
numbers.pop(0)
print(numbers)

# remove all
numbers.clear()
print(numbers)
```

# Searching in List

```python
numbers_list = [1,9 ,2, 3, 2, 4,2,6,7]

print(numbers_list.index(2))   # first occurrence of 2
print(numbers_list.count(2))  # total occurrences of 2

numbers_list.sort()          # ascending
# numbers_list.sort(reverse=True)
# numbers_list.reverse()
print(numbers_list)

```

```python
# copy the list elements
a = [1, 2, 3]
b = a          # same reference
c = a.copy()   # actual copy

print(a)
print(b)
print(c)

a.insert(1,99)

print(a)
print(b)
print(c)

print(a is b)  # True
print(a is c)  # False
```

```python
from ast import Continue
# part A
student_names = []
student_marks = []

while True:
    name = input("Enter name (q for quit): ")
    if name.lower() == 'q':
        break
    marks = input("Enter marks(0-100): ")
    if not marks.isdigit():
        print("invalid numeric value")
        continue
    marks = int(marks)
    if not (0 <= marks <= 100):
        print("values out of range")
        continue

    student_names.append(name)
    student_marks.append(marks)

print(student_names)
print(student_marks)

```

```python
# part B
# total students
print("Total students are:", len(student_names))

for name, marks in zip(student_names,student_marks):
    print(name,marks)

first_student = student_names[0]
last_student = student_names[-1]
print("First student:",first_student)
print("Last student:",last_student)

print("Ali in list:", "ali" in student_names)
```

```python
# part C
print(student_names)
print(student_marks)

for name,marks in zip(student_names,student_marks):
    if marks < 40:
        student_names.remove(name)
        student_marks.remove(marks)

for i in range(len(student_marks)):
    student_marks[i] += 5
    if student_marks[i] > 100:
        student_marks[i] = 100

print(student_names)
print(student_marks)
```

```python
# part D

max_marks = max(student_marks)
min_marks = min(student_marks)
avg_marks = sum(student_marks)/len(student_marks)

print("Maximum marks:",max_marks)
print("Minimum marks:",min_marks)
print("Average marks:",round(avg_marks,2))
# make copy for each instead of this
student_marks.sort()
print(student_marks)
# student_marks.sort(reverse=True)
print(student_marks)
# student_names.reverse()
print(student_names)
```

```python
# part E
for index, (name, marks) in enumerate(zip(student_names,student_marks)):
    print(f"Student {index+1}:")
    print(f"Name: {name}, Marks: {marks}")

result_status = []
for marks in student_marks:
    if marks >= 70:
        result_status.append("Pass")
    else:
        result_status.append("Fail")

print("Result Status:", result_status)
```

```python
# part F

student_marks_copy = student_marks.copy()
student_marks_copy.append(0)
print(student_marks)
print(student_marks_copy)
```

```python
# part G

student_marks_squares = [marks*marks for marks in student_marks]

student_marks_above_seventy = [marks for marks in student_marks if marks > 70]
# student_marks_above_seventy = []
# for marks in student_marks:
#     if marks > 70:
#         student_marks_above_seventy.append(marks)

print(student_marks_squares)
print(student_marks_above_seventy)
```
