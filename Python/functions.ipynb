# FUNCTIONS & PROGRAM STRUCTURE

This phase is extremely important for AI / ML, because:

- Models are functions

- Pipelines are functions

- Training loops are functions

- Clean code = reusable functions

###What is a Function?

A function is a named block of code that:

- Performs a task

- Can be reused

- Avoids repetition

```python
def add(a, b):
    return a + b

print(add(10, 20))
print(add(5, 7))
```

```python
# default parameters
def train_model(epochs=10):
    print("Training for", epochs, "epochs")

train_model()
train_model(50)

# order doesn't matter
def student(name, age):
    print(name, age)

student(age=21, name="Ali")

# Variable-Length Arguments (*args)
def total_sum(*numbers):
    return sum(numbers)

total_sum(1, 2, 3, 4)

# Keyword Variable Arguments (**kwargs)
def model_info(**config):
    for key, value in config.items():
        print(key, value)

model_info(lr=0.01, epochs=50, batch=32)

```

```python
# Function Scope (Local vs Global)
x = 10   # global

def test():
    x = 5    # local
    print(x)

test()
print(x)
```

##A lambda function consists of three parts:

the lambda keyword, one or more arguments, and a single expression.
Format: lambda arguments: expression

Automatic Return: The result of the expression is returned automatically without needing a return statement.

Single Expression: It can take any number of arguments but is strictly limited to exactly one expression.

```python

# Example: A lambda function that adds two numbers
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8

square = lambda x: x * x
square(5)
```

```python
# Functions as Objects
def greet():
    print("Hello")

x = greet
x()
```

```python
students = [
    {"name": "Ali", "marks": [85, 90, 78], "attendance": 92},
    {"name": "Sara", "marks": [88, 91, 95], "attendance": 96},
    {"name": "Ahmed", "marks": [70, 65, 72], "attendance": 60}
]

def calculate_average(marks):

    return sum(marks) / len(marks)

def assign_grade(avg_marks):
    if avg_marks >= 85:
        grade = "A"
    elif avg_marks >= 70:
        grade = "B"
    elif avg_marks < 70:
        grade = "C"

    return grade

def check_attendance(attendance, minimum=75):
    return attendance >= minimum

def evaluate_student(student):

    avg = calculate_average(student["marks"])
    grade = assign_grade(avg)
    eligibility = check_attendance(student["attendance"])
    return ({
        "name": student["name"],
        "average": round(avg, 2),
        "grade": grade,
        "eligible": eligibility
    })

def evaluate_students(*students):
    return [evaluate_student(student) for student in students]

def generate_report(**results):
    print("\nSTUDENT PERFORMANCE REPORT\n")
    for name,data in results.items():
        print(f"Name: {data['name']}")
        print(f"Average: {data['average']}")
        print(f"Grade: {data['grade']}")
        print(f"Eligible: {data['eligible']}")
        print("-"*20)

evaluated = evaluate_students(*students)
print(students)
print(evaluated)
evaluated_sorted = sorted(evaluated, key = lambda x: x["average"], reverse=True)

report_data = {}
for student in evaluated_sorted:
    key = student["name"]      # e.g. "Ali"
    value = student            # full student dictionary
    report_data[key] = value


print(evaluated_sorted)

generate_report(**report_data)
```
