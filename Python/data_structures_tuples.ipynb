#What is a Tuple?

A tuple is:

- Ordered

- Immutable (cannot be changed after creation)

- Can store multiple data types

- allow duplicates

###`Tuple = Read-only list`

## Why Tuples Matter (AI / ML Context)

Tuples are used when:

- Data must not change

- You want safety

- Fixed configuration values

- Dataset shapes (rows, columns)

```python
# Creating tuple
t1 = (10, 20, 30)
t2 = ("Ali", "Sara", "Ahmed")
t3 = (1, "AI", True)

# Single element
x = (5,)   # comma is required

# Accessing tuple elements
nums = (10, 20, 30)

print(nums[0])
print(nums[-1])

print(nums[0:2])
print(nums[::-1])

# this is not allowed
#nums[1] = 99  # TypeError: 'tuple' object does not support item assignment


```

### Tuple Methods (VERY FEW)

Tuples have only 2 built-in methods:

```python
# count
t = (1, 2, 2, 99)
print(t.count(2))   # 2

# index
print(t.index(99))   # 3
```

```python
# Built-in Functions on Tuples
nums = (4, 1, 9)

print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
for x in nums:
    print(x)
for i, x in enumerate(nums):
    print(i, x)
```

```python
# Tuple Packing

data = 10, 20, 30
print(data)

# Tuple Unpacking
a, b, c = data
print(a,b,c)
```

```python
# Swapping Values (Pythonic)
# No temp variable needed

x = 5
y = 10

print(x,y)
x, y = y, x
print(x,y)
```

```python
# Functions return tuples by default
def stats(nums):
    return min(nums), max(nums)

low, high = stats([10, 20, 30])

print("low",low)
print("high",high)
```

```python
# Nested tuples
matrix = (
    (1, 2),
    (3, 4)
)

print(matrix[1][0])  # 3
```

##Converting Between List & Tuple

```python
t = (1, 2, 3)
lst = list(t)
print(type(lst))

lst = [4, 5, 6]
t = tuple(lst)
print(type(t))
```

```python
student_ids = (1,2,3,4,5)
student_names = ("kashif","ali","zain","umar","ahmad")

print(len(student_ids))
print("First element:",student_ids[0],"Last element:",student_ids[-1])
print("First element:",student_names[0],"Last element:",student_names[-1])
```

```python
print("First 3 student ids:",student_ids[:3])
print("Last 2 student ids:",student_ids[-2:])
print("Reversed student ids:",student_ids[::-1])

# student_ids[1] = "rahman"
```

```python
marks = (45, 67, 67, 89, 45)
print("Count of 67:",marks.count(67))
print("Index of 89:",marks.index(89))
```

```python
# Normal for loop
print("\nStudent Names:")
for name in student_names:
    print(name)

# Using enumerate
print("\nStudent Names with Index:")
for index, name in enumerate(student_names):
    print(index, name)
```

```python
# Packing values into a tuple
subject_info = ("Artificial Intelligence", 100, 50)
# Unpacking tuple values
subject_name, total_marks, passing_marks = subject_info

print("\nSubject:", subject_name)
print("Total Marks:", total_marks)
print("Passing Marks:", passing_marks)
```

```python
id_list = list(student_ids)
id_list.append(501)
# print(id_list)
student_ids = tuple(id_list)
print(student_ids)
```

```python
# Image shape tuple (Height, Width, Channels)
image_shape = (224, 224, 3)

# Unpacking image shape
height, width, channels = image_shape

print("\nImage Height:", height)
print("Image Width:", width)
print("Image Channels:", channels)
```
