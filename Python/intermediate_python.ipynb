##1. List & Dictionary Comprehensions  
Comprehensions provide a concise way to create collections. They are generally faster than standard for loops because they are optimized at the C-level.

List Comprehensions
Syntax: [expression for item in iterable if condition]

```python
# Traditional way
squares = []
for x in range(10):
    squares.append(x**2)

# Pythonic way
squares = [x**2 for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]
```

```python
# Dictionary Comprehensions
# Syntax: {key_expr: value_expr for item in iterable}
names = ['Alice', 'Bob', 'Charlie']
# Create a map of name to name length
name_lengths = {name: len(name) for name in names}
print(name_lengths)
# Result: {'Alice': 5, 'Bob': 3, 'Charlie': 7}
```

##2. Iterators & Generators  
This is about Memory Efficiency.

Iterators: Objects that can be iterated upon (using a __next__ method).

Generators: A special type of iterator defined using a function and the yield keyword. Unlike lists, they don't store all values in memory; they generate them on the fly.

```python
# A simple generator
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Pauses here and returns value
        count += 1

counter = count_up_to(5)
print(next(counter)) # 1
print(next(counter)) # 2
```

##3. map(), filter(), and reduce()
These are functional programming tools.

map(func, iterable): Applies a function to every item in an object.

filter(func, iterable): Keeps items where the function returns True.

reduce(func, iterable): (From functools module) Continually applies a function to pairs of elements to reduce the list to a single value.

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Double everything
doubled = list(map(lambda x: x * 2, nums))

# Only numbers > 2
filtered = list(filter(lambda x: x > 2, nums))

# Sum of all numbers
total = reduce(lambda x, y: x + y, nums)

print(doubled)
print(filtered)
print(total)
```

##4. Decorators
A decorator is a function that takes another function and extends its behavior without explicitly modifying it. Think of it as "wrapping" a gift.

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

##5. The datetime Module
Python doesn't have a "date" data type of its own, so we import the datetime module to work with dates as objects.

###Key Classes:  
datetime.date: Year, month, day.

datetime.time: Hour, minute, second, microsecond.

datetime.datetime: Combination of date and time.

datetime.timedelta: A duration expressing the difference between two dates.

###Essential Methods:  
datetime.now(): Current local date and time.

strftime(): (String Format Time) Formats objects into readable strings.

strptime(): (String Parse Time) Parses a string into a datetime object.

```python
from datetime import datetime, timedelta

now = datetime.now()
print(f"Current Time: {now}")

# Formatting: %Y = Year, %m = Month, %d = Day
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print(formatted)
# Math with dates
future_date = now + timedelta(days=10)
print(f"10 days from now: {future_date}")
```

Part 1: Concepts
Generators use "Lazy Evaluation." They only hold one item in memory at a time and "yield" it when requested. A List must store all 1 million items in RAM simultaneously, which can lead to a MemoryError.

map() changes the form of every item (e.g., squaring every number). filter() changes the quantity of items by keeping only those that meet a condition (e.g., keeping only even numbers).

The @ symbol is syntactic sugar. @my_decorator followed by def func(): is equivalent to writing func = my_decorator(func). It replaces the original function with the "wrapped" version.

%m represents the Month (01-12), while %M represents Minutes (00-59).
