#What is a Set?

A set is:

- Unordered

- Mutable

- Stores only unique values

- Does NOT allow duplicates

###`A bag of unique items with no order`

###Why Sets Matter (AI / ML)

Used for:

- Removing duplicate data

- Feature uniqueness

- Fast membership checks

- Mathematical set operations

```python
# creating sets
numbers = {1, 2, 3, 4}
names = {"Ali", "Sara", "Ahmed"}
mixed = {1, "AI", True}

# creating empyty set
s = {}   # this creates a dictionary
print(type(s))
s = set()
print(type(s))

# Automatic Duplicate Removal
nums = {1, 2, 2, 3, 3, 3, 4}
print(nums)   # {1, 2, 3} - does not guarantee order

```

```python
# Accessig elements
for x in nums:
    print(x)
# nums[0]   # error
```

## Adding Elements

```python
# adding single element
nums.add(5)
# adding multiple elements
nums.update([6,7,8])
print(nums)
```

## Removing Elements

```python
# nums.remove(3) # error if element not found

nums.discard(10) # safe to remove element
print(nums)
# nums.pop() # removes random error if empty
print(nums)

# nums.clear() # removes all
print(nums)

```

```python
print(2 in nums)
print(10 not in nums)
```

##Set Mathematical Operations

```python
# UNION
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
print(a.union(b))
```

```python
# INTERSECTION
print(a & b)
print(a.intersection(b))
```

```python
# SYMMETRIC DIFFERENCE - KEEP UNIQUE FROM EACH SET
print(a ^ b)
print(a.symmetric_difference(b))
```

```python
# SUBSET AND SUPERSET
a = {1, 2}
b = {1, 2, 3}

print(a.issubset(b))
print(a.issuperset(b))
```

```python
# copying sets
s1 = {1, 2, 3}
s2 = s1.copy()
```

###Frozen Sets (Immutable Sets)

- Cannot add/remove

- Can be dictionary key

```python
fs = frozenset([1, 2, 3])
# fs.add(5) - error
print(fs)
```

##Built-in Functions

```python
nums = {5, 1, 9}

print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
squares = {x * x for x in range(5)}
print(squares)
```

```python
# Converting Between Types
lst = [1, 2, 2, 3]
unique = set(lst)
back_to_list = list(unique)

print(unique)
```

```python
set_a = {1, 2, 3, 4,5,6}
set_b = {3, 4, 5, 6}

print("Union:", set_a.union(set_b))
print("Intersection:", set_a.intersection(set_b))
print("Difference:", set_a.difference(set_b))
print("Symmetric difference:", set_a.symmetric_difference(set_b))

print("Set a is subset of set b:",set_a.issubset(set_b))
print("Set a is superset of set b:",set_a.issuperset(set_b))
```

```python
set_a_copy = set_a.copy()
print(set_a_copy)
set_a_copy.clear()
print(set_a_copy)
print(set_a)
```
